```python
import numpy as np
from IPython.display import Audio


def generate_wave(frequency = 440, duration = 3.0, sample_rate = 44100):
    t = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)
    y = 0.5 * np.sin(2 * np.pi * frequency * t)

    print(f"Frequency = {frequency} Hz")
    return Audio(y, rate=sample_rate)
```

```python
generate_wave(440)
```

```python
generate_wave(660)
```

```python
generate_wave(880)
```

```python
generate_wave(1100)
```
